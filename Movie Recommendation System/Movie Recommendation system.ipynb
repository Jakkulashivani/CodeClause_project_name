# importing libraries

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import difflib
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

# loading data set

```python
df=pd.read_csv("Downloads\movies.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
selecting_features=['genres','keywords','tagline','cast','director']
```

```python
selecting_features
```

```python
for feature in selecting_features:
    df[feature]=df[feature].fillna(" ")
```

```python
new_df=df['genres']+' '+df['keywords']+''+df['tagline']+''+df['cast']+''+df['director']
```

```python
new_df
```

# feature extraction

```python
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(new_df)
```

```python
feature_vectors
```

# getting similarity scores using cosine _similarity

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

# getting the movie name from user

```python
movie_name=input("enter the movie name:")
```

# creating the list with all the movie names given in the dataset

```python
list_all_titles=df['title'].tolist()
```

```python
list_all_titles
```

# finding the close match for the movie name given by the user

```python
find_close_match=difflib.get_close_matches(movie_name,list_all_titles)
```

```python
find_close_match
```

```python
close_match=find_close_match[0]
```

```python
close_match
```

# finding the index of the movie with tittle

```python
index_of_the_movie=df[df.title==close_match]['index'].values[0]
```

```python
index_of_the_movie
```

# getting a list of similar movies

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
similarity_score
```

```python
len(similarity_score)
```

```python
#sorting the movies based on similarity score
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
sorted_similar_movies
```

```python
#print the names of similar movie base on index of the movie
```

```python
print("movie suggested for you:\n")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
        i=i+1
```
